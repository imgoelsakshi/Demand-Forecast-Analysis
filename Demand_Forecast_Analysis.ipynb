## 課題背景

電力会社の需給管理部門では、電力の需要と供給を1時間単位で調整する必要があります。  
この需給予測の精度が低いと、以下のようなリスクが高まります。

- インバランス料金の発生
- 収益の不安定化

現状、需給管理部門では **年間約5億円** のインバランス料金が発生していると仮定されます（本課題で定めた簡易試算に基づく）。
また現状の予測方法は、担当者が予測日の気象条件やカレンダー情報、経験に基づいて。過去の類似日を複数取得し、その類似日の需要実績の平均値を使用しています。

### インバランス料金とは

需給計画と実績の差（インバランス）に対して、送配電事業者が調整した電力量に応じて課される料金です。

- **不足インバランス**：計画より実需が上回った場合、不足分を送配電事業者が補填し、その費用が課されます。
- **余剰インバランス**：実需が計画を下回った場合、余剰分は買い取られますが、一定のペナルティが課されることがあります。

#### 本課題で使用する簡易インバランス料金モデル

本課題では以下のように簡略化したインバランス料金モデルを用います：

```
インバランス料金（円）＝ ｛|予測電力量 − 実績電力量|（万kWh）｝² × 0.8
```

# **2. What Are Imbalance Fees?**

Imbalance fees are charges applied by the transmission/distribution operator when there is a gap between the planned and actual electricity demand.

### **Shortage Imbalance**

* Actual demand > Planned demand
* Shortage is supplied by the operator
* Additional cost is charged

### **Surplus Imbalance**

* Actual demand < Planned demand
* Excess is bought by the operator
* Penalties may apply

## **Simplified Imbalance Fee Model (Used in This Assignment)**

```
Imbalance Fee (yen) = ( |Forecasted Power − Actual Power| in 10,000 kWh units )² × 0.8
```

## Merging all weather files

````python
# List of 7 weather files
weather_files = [
      "大阪.csv",
      "京都.csv",
      "和歌山.csv",
      "彦根.csv",
      "潮岬.csv",
      "神戸.csv",
      "豊岡.csv"
  ]
````

````python
# load each csv, add city column, store in a list

weather_dfs = []
for file in weather_files:
  city_name = file.split('.')[0] # using file name as city name
  df = pd.read_csv(file, parse_dates=[0])
  df['city'] = city_name # creating a column to add the city name
  weather_dfs.append(df) # adding all the dataframes to a list
````

````python
# concatenate all cities dataframes into one long dataframe

weather_all = pd.concat(weather_dfs, ignore_index=True)
````

````python
weather_all.head()
````

````python
#rename columns
weather_all = weather_all.rename(columns= {'降水量':'precipitation', '気温':'temp', '露点温度':'dew_point', '湿度':'humidity', '風速':'wind_speed', '風向':'wind_direction', '降雪':'snowfall'})
````

````python
weather_all.head()
````

````python
# dropping wind direction and city name column since it cannot be combined across all 7 cities

weather_all = weather_all.drop(['wind_direction', 'city'], axis=1)
````

````python
weather_all.head()
````

````python
# aggregate weather by datetime (average across cities)

weather_agg = weather_all.groupby('datetime').mean().reset_index()
````

````python
weather_agg.head()
````

````python
weather_agg.describe()
````

````python
# load demand file
demand = pd.read_csv('電力需要.csv', parse_dates=[0])
demand = demand.rename(columns={'当日実績(万kW)': 'demand_mankw'})
````
